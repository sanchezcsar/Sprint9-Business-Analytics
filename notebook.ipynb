2.1.3 ¿Cuál es la duración de cada sesión?

```python
# Paso 2.1.4: ¿Con qué frecuencia los usuarios regresan?
first_session = visits.groupby('Uid')['Start Ts'].min().reset_index()
first_session.columns = ['Uid', 'first_ts']
visits = visits.merge(first_session, on='Uid')
visits['days_since_first'] = (visits['Start Ts'] - visits['first_ts']).dt.days

# Contar cuántos usuarios regresaron al menos una vez en los siguientes días
repeat_visits = visits[visits['days_since_first'] > 0]
user_return_frequency = repeat_visits.groupby('Uid')['days_since_first'].min()

# Mostrar resultados
print("Frecuencia con la que los usuarios regresan (días desde la primera visita):")
display(user_return_frequency.describe())
```

2.2.3 ¿Cuál es el tamaño promedio de compra?
Para calcular esta métrica, se utilizó el ingreso total dividido entre el número total de pedidos. Este valor nos da una idea clara del ingreso que se genera por cada transacción, lo cual es fundamental para estimar ingresos futuros y planificar campañas de adquisición de usuarios.

```python
# Paso 2.2.4: ¿Cuánto dinero traen? (LTV)
# Obtener la primera compra de cada usuario
first_order_month = orders.groupby('Uid')['order_month'].min().reset_index()
first_order_month.columns = ['Uid', 'first_purchase_month']

# Unir al dataframe de órdenes
orders_with_cohort = pd.merge(orders, first_order_month, on='Uid')

# Calcular edad de la cohorte
orders_with_cohort['cohort_age'] = (orders_with_cohort['order_month'] - orders_with_cohort['first_purchase_month']).apply(lambda x: x.n)

# Calcular número de usuarios por cohorte
cohort_sizes = orders_with_cohort.groupby('first_purchase_month')['Uid'].nunique().reset_index()
cohort_sizes.columns = ['first_purchase_month', 'n_buyers']

# Calcular ingresos por cohorte y edad
ltv_data = orders_with_cohort.groupby(['first_purchase_month', 'cohort_age']).agg({'Revenue': 'sum'}).reset_index()
ltv_data = pd.merge(ltv_data, cohort_sizes, on='first_purchase_month')
ltv_data['ltv'] = ltv_data['Revenue'] / ltv_data['n_buyers']

# Tabla pivote acumulada
ltv_pivot = ltv_data.pivot_table(index='first_purchase_month', columns='cohort_age', values='ltv', aggfunc='sum').cumsum(axis=1)

# Visualización
plt.figure(figsize=(12, 6))
sns.heatmap(ltv_pivot, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('LTV acumulado por cohorte y edad')
plt.xlabel('Meses desde la primera compra')
plt.ylabel('Mes de primera compra')
plt.tight_layout()
plt.show()

ltv_pivot
```

```python
# Paso 2.3.2: Cálculo del CAC

# Obtener la fuente de tráfico de la primera visita por usuario
first_sources = visits.sort_values(by='Start Ts').drop_duplicates('Uid', keep='first')[['Uid', 'Source Id', 'Start Ts']]
first_sources.columns = ['Uid', 'first_source', 'first_visit_ts']

# Obtener la fecha de primera compra
first_orders = orders.groupby('Uid')['Buy Ts'].min().reset_index()
first_orders['first_order_month'] = first_orders['Buy Ts'].astype('datetime64[M]')

# Unir visitas y órdenes para tener fuente y fecha de compra
user_sources = pd.merge(first_sources, first_orders, on='Uid', how='inner')

# Calcular el número de compradores por fuente y mes
buyers_by_source_month = user_sources.groupby(['first_source', 'first_order_month'])['Uid'].nunique().reset_index()
buyers_by_source_month.columns = ['source_id', 'month', 'n_buyers']

# Sumar gastos de marketing por fuente y mes
monthly_costs_by_source = costs.groupby(['source_id', 'month'])['costs'].sum().reset_index()

# Calcular el CAC
cac = pd.merge(buyers_by_source_month, monthly_costs_by_source, on=['source_id', 'month'], how='inner')
cac['cac'] = cac['costs'] / cac['n_buyers']

# Mostrar resultados
display(cac.head())
```

En esta sección calculamos el Costo de Adquisición de Clientes (CAC) por fuente. Para ello, se divide el costo total por fuente entre el número de usuarios únicos que realizaron al menos una compra provenientes de esa fuente. Esta métrica es clave para evaluar la eficiencia de cada canal de marketing y determinar qué tan costoso es adquirir un nuevo cliente desde cada fuente de tráfico.

```python
# Paso 2.3.3: Cálculo del ROMI por fuente
romi_list = []
for source in user_sources['first_source'].dropna().unique():
    source_users = user_sources[user_sources['first_source'] == source]
    source_orders = pd.merge(source_users[['Uid', 'first_source', 'first_order_month']], orders, on='Uid')
    source_orders['cohort_age'] = (
    source_orders['Buy Ts'].dt.to_period('M') - source_orders['first_order_month'].dt.to_period('M')
).apply(lambda x: x.n).astype(int)

    cohort_sizes = source_users.groupby('first_order_month')['Uid'].nunique().reset_index()
    cohort_sizes.columns = ['first_order_month', 'n_buyers']
    ltv = source_orders.groupby(['first_order_month', 'cohort_age']).agg({'Revenue': 'sum'}).reset_index()
    ltv = pd.merge(ltv, cohort_sizes, on='first_order_month', how='left')
    ltv['ltv'] = ltv['Revenue'] / ltv['n_buyers']

    # Agrupar correctamente los costos por mes antes del merge
    cost = costs[costs['source_id'] == source].groupby('month')['costs'].sum().reset_index()
    cost.columns = ['first_order_month', 'costs']

    romi = pd.merge(ltv, cost, on='first_order_month', how='left')
    romi = romi.dropna(subset=['ltv', 'costs'])
    romi['cac'] = romi['costs'] / romi['n_buyers']
    romi['romi'] = romi['ltv'] / romi['cac']
    romi['source_id'] = source
    romi_list.append(romi)

romi_data = pd.concat(romi_list)
romi_pivot = romi_data.pivot_table(index='source_id', columns='cohort_age', values='romi', aggfunc='mean')

# Mostrar tabla ROMI
display(romi_pivot.round(2))

# Visualización con heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(romi_pivot, annot=True, fmt='.2f', cmap='YlOrBr')
plt.title('ROMI acumulado por fuente y edad de cohorte')
plt.xlabel('Meses desde la primera compra')
plt.ylabel('Fuente de adquisición')
plt.tight_layout()
plt.show()
```
